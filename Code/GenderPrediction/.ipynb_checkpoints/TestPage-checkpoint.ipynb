```python
import warnings
warnings.filterwarnings('ignore')
from tkinter import *
import sys
import math
import tkinter as tk
from tkinter import ttk
import pandas as pd
import pylab as pl
from tkinter import filedialog
from pydub import AudioSegment
from pydub.playback import play
import pygame
import wave
mainFrame= tk.Tk()
mainFrame.geometry("890x700")
mainFrame.title("Welcome to the System!!!")

viewerPane = Frame(mainFrame)
viewerPane.grid(row=0,column=0,rowspan=10,columnspan=4, sticky="NESW")
viewerPane.grid_columnconfigure(0, weight=1)
viewerPane.grid_rowconfigure(0, weight=1)

actionPane = Frame(mainFrame)
actionPane.grid(row=11,column=0,rowspan=30,columnspan=4, sticky="NESW")

myLabel1 = Label(viewerPane, text="Split the DataSet!",fg="blue",height=3,width=55,font=('Timesnewsroman', 10))
myLabel2 = Label(viewerPane, text="Training %",fg="blue",height=3,width=50,font=('Timesnewsroman', 10))
myLabel1.grid(row=1, column=0)
myLabel2.grid(row=1, column=1)

entryTrain = tk.Entry(viewerPane, width=10)
entryTrain.grid(column=1, row=1, padx=10, pady=5, sticky=tk.E)

myLabel3 = Label(viewerPane, text = "Choose Classification Model!",fg="blue",height=3,width=55,font=('Timesnewsroman', 10))
myLabel3.grid(row=3,column=0,)

combostyle = ttk.Style()
comboModel = ttk.Combobox(viewerPane, values=[
                                    "Artificial Neural Network", 
                                    "Support Vector Machine",
                                    "Random Forest",
                                    "Logistic Regression",
                                    "Naive Bayes Classifier"],width=25,height=10,font=('Timesnewsroman', 10))
comboModel.grid(row=3,column=1,padx=10, pady=10 )
comboModel.current(0)

button=Button(viewerPane, text="OK",fg='blue',command=lambda: result())
button.grid(column=0, row=5,columnspan=2,padx=10, pady=10)

ttk.Separator(viewerPane, orient=HORIZONTAL).grid(column=0, row=10, columnspan=45, sticky='ew')


from ipynb.fs.full.Test import *
from ipynb.fs.full.classificationbyANN import *
def result():
    global entryTrain
    global comboModel
    global train,test,confusion,report
    
    content = entryTrain.get()
    content=1-(int(float(content))/100)
    chooseModel=comboModel.get()
  
    if chooseModel =='Random Forest':
        train,test,confusion,report,error=RF(content)
        display(train,test,confusion,report,error)
        
    elif chooseModel =='Artificial Neural Network':
        train,test,confusion,report,error=ANN(content)
        display(train,test,confusion,report,error)
        
    elif chooseModel =='Support Vector Machine':
        train,test,confusion,report,error=SVM(content)
        display(train,test,confusion,report,error)
        
    elif chooseModel =='Logistic Regression':
        train,test,confusion,report,error=LR(content)
        display(train,test,confusion,report,error)
        
    else:
        train,test,confusion,report,error=GNB(content)
        display(train,test,confusion,report,error)
        
def display(train,test,matrix,report,error):
    trainLabel1 = Label(actionPane,text = "Model Accuracy for Training:",height=3,width=55,font=('Timesnewsroman', 10))
    trainLabel1.grid(column=0, row=13,sticky=tk.E)
    trainLabel2=Label(actionPane,text=train)
    trainLabel2.grid(column=1,row=13,sticky=tk.W)
    testLabel1 = Label(actionPane,text = "Model Accuracy for Testing:",height=3,width=55,font=('Timesnewsroman', 10))
    testLabel1.grid(column=0, row=17,sticky=tk.E)
    testLabel2=Label(actionPane,text=test)
    testLabel2.grid(column=1,row=17,sticky=tk.W)
    matrixLabel1= Label(actionPane,text = "Confusion Matrix:",height=3,width=55,font=('Timesnewsroman', 10))
    matrixLabel1.grid(column=0,row=21,sticky=tk.E)
    matrixLabel2=Label(actionPane,text=matrix)
    matrixLabel2.grid(column=1,row=21,sticky=tk.W)
    reportLabel1= Label(actionPane,text = "Classification Report:",height=3,width=55,font=('Timesnewsroman', 10))
    reportLabel1.grid(column=0,row=25,sticky=tk.E)
    reportLabel2=Label(actionPane,text=report)
    reportLabel2.grid(column=1,row=25,sticky=tk.W)
    errorLabel1= Label(actionPane,text = "Error Rate/Missclassification Rate:",height=3,width=55,font=('Timesnewsroman', 10))
    errorLabel1.grid(column=0,row=30,sticky=tk.E)
    errorLabel2=Label(actionPane,text=error)
    errorLabel2.grid(column=1,row=30,sticky=tk.W)
    
mainFrame.resizable(False, False)
mainFrame.mainloop()
```
