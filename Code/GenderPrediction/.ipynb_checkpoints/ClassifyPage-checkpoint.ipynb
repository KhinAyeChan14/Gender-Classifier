```python
import warnings
warnings.filterwarnings('ignore')
from tkinter import *
import tkinter as tk
from tkinter import ttk
from tkinter import filedialog
import pygame
import wave
import import_ipynb
import mfcc
from mfcc import compute_mfcc
import librosa
import scipy.io.wavfile as wav
from scipy.io import wavfile
from python_speech_features import mfcc
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.externals import joblib

mainFrame= tk.Tk()
mainFrame.geometry("800x500")
mainFrame.title("Welcome from Classify Page!!!")

title = Label(mainFrame, text="What is Your Voice Gender? ",fg="blue",height=3,width=55,font=('Timesnewsroman', 10))
title.grid(column=1, row=1,padx=5, pady=5)

button1=Button(mainFrame, text="Browse",fg='blue',command=lambda: chooseFile())
button1.grid(column=0, row=3,padx=5, pady=5)
 
button2=Button(mainFrame, text="Play",fg='blue',command=lambda: masker_screen())
button2.grid(column=0, row=5)

playLabel= Label(mainFrame, text="Play the chosen wave file! ",fg="red",height=3,width=55,font=('Timesnewsroman', 10))
playLabel.grid(column=1, row=5,padx=5, pady=5,columnspan=2)


button3=Button(mainFrame, text="Classify",fg='blue',command=lambda: classifyGender())
button3.grid(column=0, row=7,padx=5, pady=5)


from ipynb.fs.full.Classify import *
import pickle
def chooseFile():
    
    global audio_file_path
    audio_file_path = filedialog.askopenfilename(filetypes=(("Audio Files", ".wav"), ("All Files", "*.*")))
    filename = Label(mainFrame,text = audio_file_path,height=3,width=75,font=('Timesnewsroman', 10))
    filename.grid(row=3,column=1,columnspan=2)
   
    
def masker_screen():  
    # we will also use the audio_file_name global variable
    global m_screen, audio_file_path 

    m_screen = Toplevel()
    m_screen.geometry('150x130')
    m_screen.transient(mainFrame) 
    m_label = Label(m_screen, text = "Playing Sound")
    m_label.pack(anchor = CENTER)

    if audio_file_path: # play sound if just not an empty string
        file_wav = wave.open(audio_file_path)
        frequency = file_wav.getframerate()
        pygame.mixer.init(frequency=frequency)
        pygame.mixer.music.load(audio_file_path)
        pygame.mixer.music.play()
        
def classifyGender():

    model= Label(mainFrame, text='Model',fg="blue",height=3,width=55,font=('Timesnewsroman', 10))
    model.grid(column=1, row=7,padx=5, pady=5,rowspan=3,columnspan=1)

    label= Label(mainFrame, text='Label',fg="blue",height=3,width=55,font=('Timesnewsroman', 10))
    label.grid(column=2, row=7,padx=5, pady=5,rowspan=3,columnspan=1)
    
    df = pd.DataFrame()
    write_mfeatures=[]
    sampling_rate,speech_signal = wavfile.read(audio_file_path)
    mfcc_feat=compute_mfcc(speech_signal,sampling_rate)
    write_feat=mfcc_feat.tolist()[0]
    
    #loaded_model = pickle.load(open('RF_model.sav', 'rb'))
    loaded_model = joblib.load('RF_model.sav')
    label1= loaded_model.predict([write_feat])

    if label1[0] == 1:
        gender1='male'
    else:
        gender1='female'
    #print(gender1)
    model1= Label(mainFrame, text='Random Forest',fg="blue",height=3,width=55,font=('Timesnewsroman', 10))
    model1.grid(column=1, row=12,padx=5, pady=5,rowspan=3,columnspan=1)
    result1= Label(mainFrame, text=gender1,fg="blue",height=3,width=55,font=('Timesnewsroman', 10))
    result1.grid(column=2, row=12,padx=5, pady=5,rowspan=3,columnspan=1)
    
    #loaded_model = pickle.load(open('LR_model.sav', 'rb'))
    loaded_model = joblib.load('LR_model.sav')
    label2= loaded_model.predict([write_feat])

    if label2[0] ==1:
        gender2='male'
    else:
        gender2='female'
    #print(gender2)
    model2= Label(mainFrame, text='Logistic Regression',fg="blue",height=3,width=55,font=('Timesnewsroman', 10))
    model2.grid(column=1, row=17,padx=5, pady=5,rowspan=3,columnspan=1)
    result2= Label(mainFrame, text=gender2,fg="blue",height=3,width=55,font=('Timesnewsroman', 10))
    result2.grid(column=2, row=17,padx=5, pady=5,rowspan=3,columnspan=1)
    
    #loaded_model = pickle.load(open('GNB_model.sav', 'rb'))
    loaded_model = joblib.load('GNB_model.sav')
    label3= loaded_model.predict([write_feat])

    if label3 [0]==1:
        gender3='male'
    else:
        gender3='female'
    #print(gender3)
    model3= Label(mainFrame, text='Gaussian Naive Bayes',fg="blue",height=3,width=55,font=('Timesnewsroman', 10))
    model3.grid(column=1, row=23,padx=5, pady=5,rowspan=3,columnspan=1)
    result3= Label(mainFrame, text=gender3,fg="blue",height=3,width=55,font=('Timesnewsroman', 10))
    result3.grid(column=2, row=23,padx=5, pady=5,rowspan=3,columnspan=1)
    
    #loaded_model = pickle.load(open('SVM_model.sav', 'rb'))
    loaded_model = joblib.load('SVM_model.sav')
    label4= loaded_model.predict([write_feat])
    
    if label4[0] ==1:
        gender4='male'
    else:
        gender4='female'
    #print(gender4)
    model4= Label(mainFrame, text='Support Vector Machine',fg="blue",height=3,width=55,font=('Timesnewsroman', 10))
    model4.grid(column=1, row=28,padx=5, pady=5,rowspan=3,columnspan=1)
    result4= Label(mainFrame, text=gender4,fg="blue",height=3,width=55,font=('Timesnewsroman', 10))
    result4.grid(column=2, row=28,padx=5, pady=5,rowspan=3,columnspan=1)
    
    min_max_scaler = preprocessing.MinMaxScaler()
    X_scale = min_max_scaler.fit_transform([write_feat])
    #loaded_model = pickle.load(open('ANN_model.sav', 'rb'))
    loaded_model = joblib.load('ANN_model.sav')
    label5= loaded_model.predict([X_scale])
    label5= [ 1 if y>=0.5 else 0 for y in label5]

    if label5[0] ==1:
        gender5='male'
    else:
        gender5='female'
    #print(gender5)
    model5= Label(mainFrame, text='Artificial Neural Network',fg="blue",height=3,width=55,font=('Timesnewsroman', 10))
    model5.grid(column=1, row=33,padx=5, pady=5,rowspan=3,columnspan=1)
    result5= Label(mainFrame, text=gender5,fg="blue",height=3,width=55,font=('Timesnewsroman', 10))
    result5.grid(column=2, row=33,padx=5, pady=5,rowspan=3,columnspan=1)
    

mainFrame.resizable(False, False)
mainFrame.mainloop() 
```
