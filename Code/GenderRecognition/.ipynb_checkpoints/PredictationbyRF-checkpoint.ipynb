```python
import numpy as np
import pandas as pd
from sklearn import preprocessing, neighbors
from sklearn.cross_validation import train_test_split
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('features.csv')
df.head(5)
```

```python
from sklearn.linear_model import LogisticRegression
#producing x and y
x= pd.read_csv('features.csv',header=None,usecols=range(1,12) ,skiprows=1)
x = np.array(x)
print(x)
```

```python
y= pd.read_csv('features.csv',usecols=range(12,13),header=None,skiprows=1)
y = np.array(y)
print(y)

#lABEL ENCODING
le = preprocessing.LabelEncoder()
le.fit(y)
y = le.transform(y)
print(y)
```

```python
#Dividing the data randomly into training and test set
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)

from sklearn.ensemble import RandomForestClassifier
rf= RandomForestClassifier( )
rf.fit(x_train, y_train)

print('Accuracy1 for RandomForest :',rf.score(x_train,y_train))
print('Accuracy2 for RandomForest  :',rf.score(x_test,y_test))
```

```python
y_pred_class = rf.predict(x_test)
print(y_pred_class)
print(y_test)
from sklearn import metrics
print('Classification Accuracy is:',metrics.accuracy_score(y_test, y_pred_class))
```

```python
# calculate the percentage of ones
# because y_test only contains ones and zeros, we can simply calculate the mean = percentage of ones
print('Percentage of ones:',y_test.mean())
# calculate the percentage of zeros
print('percentage of zeros:',1 - y_test.mean())
```

```python
# calculate null accuracy in a single line of code
# only for binary classification problems coded as 0/1
print(max(y_test.mean(), 1 - y_test.mean()))
'''This means that a dumb model that always predicts 0 would be right 68% of the time. This shows how classification accuracy is not that good as it's close to a dumb model Its a good way to know the minimum we should achieve with our models.'''
```

```python
# print the first 25 true and predicted responses
print('True:', y_test[0:25])
print('False:', y_pred_class[0:25])
```

```python
# IMPORTANT: first argument is true values, second argument is predicted values
# this produces a 2x2 numpy array (matrix)
print(metrics.confusion_matrix(y_test, y_pred_class))
# save confusion matrix and slice into four pieces
confusion = metrics.confusion_matrix(y_test, y_pred_class,)
exp_series = pd.Series(y_test)
pred_series = pd.Series(y_pred_class)
pd.crosstab(y_test, y_pred_class, rownames=['Actual'], colnames=['Predicted'],margins=True)
```

```python
plt.matshow(confusion)
plt.title('Confusion Matrix Plot')
plt.colorbar()
plt.xlabel('Precited')
plt.ylabel('Actual')
plt.show();
```

```python
TN=10
FP=14
FN=1
TP=88
classification_error = (FP + FN) / float(TP + TN + FP + FN)
print(classification_error)
print(1 - metrics.accuracy_score(y_test, y_pred_class))
```

```python
sensitivity = TP / float(FN + TP)
print(sensitivity)
print(metrics.recall_score(y_test, y_pred_class))
```

```python
specificity = TN / (TN + FP)
print(specificity)
```

```python
false_positive_rate = FP / float(TN + FP)

print(false_positive_rate)
print(1 - specificity)
```

```python
precision = TP / float(TP + FP)

print(precision)
print(metrics.precision_score(y_test, y_pred_class))
```
