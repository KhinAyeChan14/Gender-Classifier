```python
import import_ipynb
import pandas as pd
import os
import mfcc
from mfcc import compute_mfcc
import librosa
import scipy.io.wavfile as wav
from scipy.io import wavfile
```

```python
df = pd.DataFrame()
print('Extracting features')

label_col=['label']
mfcc_col=['mfcc'+str(i+1) for i in list(range(12))]
col = mfcc_col+label_col
label1='male'
label2='female'

directory1=os.listdir("/home/khin/Male/")
directory2=os.listdir("/home/khin/Female/")


for wav_file in directory1:
    write_mfeatures=[]
    sampling_rate,speech_signal = wavfile.read('/home/khin/Male/'+wav_file)   
    mfcc_features=compute_mfcc(speech_signal,sampling_rate)
    write_mfeatures=mfcc_features.tolist()[0]+[label1]
    df = df.append([write_mfeatures], ignore_index=True)
    
for wav_file in directory2:
    write_ffeatures=[]
    sampling_rate,speech_signal = wavfile.read('/home/khin/Female/'+wav_file)   
    mfcc_features=compute_mfcc(speech_signal,sampling_rate)
    write_ffeatures=mfcc_features.tolist()[0]+[label2]
    df = df.append([write_ffeatures], ignore_index=True)
    
df.columns = col
df.to_csv('features.csv', index=False)

pd.read_csv('features.csv', delimiter = ',')
```
